```python
import matplotlib.pyplot as plt
import pandas as pd
from ipywidgets import interact, widgets
```

```python
# Create a scatter plot function
def dynamic_scatter_plot(x_col, y_col):
    plt.figure(figsize=(10, 6))
    plt.scatter(mov[x_col], mov[y_col], c=mov['Rating'], cmap='viridis', alpha=0.6)
    plt.colorbar(label='Rating')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.title(f'Scatter Plot of Movies ({x_col} vs. {y_col})')
    plt.show()

# Create interactive widgets for selecting columns
x_column_widget = widgets.Dropdown(options=mov.columns, description='X-axis:')
y_column_widget = widgets.Dropdown(options=mov.columns, description='Y-axis:')

# Create an interactive plot using ipywidgets
interact(dynamic_scatter_plot, x_col=x_column_widget, y_col=y_column_widget)
```
